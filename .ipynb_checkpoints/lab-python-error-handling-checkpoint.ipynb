   - If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, display an error message and ask them to re-enter the product name. *Hint: you will need to pass inventory as a parameter*
   - Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.

4. Test your code by running the program and deliberately entering invalid quantities and product names. Make sure the error handling mechanism works as expected.


````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]


def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_input = False
        while not valid_input:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity >= 0:
                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")
            except ValueError:
                print("Invalid input. Please enter a valid quantity.")
    return inventory


def get_customer_orders(products, inventory):
    while True:
        try:
            num_orders = int(input("Enter the number of customer orders: "))
            if num_orders < 0:
                raise ValueError("The number of orders cannot be negative.")
            break
        except ValueError as e:
            print(f"Invalid. Please re-enter the number of orders")
    
    customer_orders = set()
    while len(customer_orders) < num_orders:
        try:
            product_order = input("Enter the name of a product that a customer wants to order: ")
            if product_order not in products:
                raise ValueError(f"{product_order} is not available in the inventory. Please re-enter the product name.")
            if inventory.get(product_order, 0) <= 0:
                raise ValueError(f"{product_order} is out of stock.")
            customer_orders.add(product_order)
            if len(customer_orders) == num_orders:
                break
        except ValueError as e:
            print(e)
    
    return customer_orders


def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if product in inventory and inventory[product] > 0:
            inventory[product] -= 1
        else:
            print(f"Insufficient quantity for {product}" if product in inventory else f"Product {product} not found in inventory.")
    inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}

    return inventory
            
            
def calculate_total_price(customer_orders):
    total_price = 0
    for product in customer_orders:
        while True:
            try:
                price_input = input(f"Enter the price of {product}: ")
                price = float(price_input)
                if price < 0:
                    raise ValueError("Price cannot be negative.")
                total_price += price
                break
            except ValueError as e:
                print(f"Invalid price. Please  re-enter the price.")
    return total_price


def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_of_products_ordered = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_of_products_ordered

def print_order_statistics(order_statistics):
    total_products_ordered, percentage_of_products_ordered = order_statistics
    print(f"\nOrder Statistics:\nTotal Products Ordered: {total_products_ordered}\nPercentage of Products Ordered: {percentage_of_products_ordered:.2f}%\n")

def print_updated_inventory(inventory):
    print("Updated Inventory:")
    for product, quantity in inventory.items():
        print(f"{product}: {quantity}")

        


inventory = initialize_inventory(products)
print()
customer_orders = get_customer_orders(products, inventory) 
order_statistics = calculate_order_statistics(customer_orders, products)
total_products_ordered, percentage_of_products_ordered = order_statistics
print(f"\nOrder Statistics:\nTotal Products Ordered: {total_products_ordered}")
print(f"Percentage of Products Ordered: {percentage_of_products_ordered:.2f}%")
print()
inventory = update_inventory(customer_orders, inventory)
print()
print_updated_inventory(inventory)
total_price = calculate_total_price(customer_orders)
print(f"The total price of your order is: ${total_price:.2f}")
````
